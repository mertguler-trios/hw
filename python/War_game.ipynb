```python
import random

suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 
            'Nine':9, 'Ten':10, 'Jack':11, 'Queen':12, 'King':13, 'Ace':14}
```

```python
class Card:

    def __init__(self,suit,rank):
        self.suit = suit.capitalize()
        self.rank = rank.capitalize()
        self.value = values[rank.capitalize()]

    def __str__(self):
        return self.rank + ' of ' + self.suit
```

```python
class Deck:

    def __init__(self):

        self.all_cards = []

        for suit in suits:
            for rank in ranks:
                self.all_cards.append(Card(suit,rank))

    def shuffle(self):

        random.shuffle(self.all_cards)

    def draw_one(self):

        return self.all_cards.pop()
```

```python
class Player:

    def __init__(self,name):
        self.name = name
        self.all_cards = []

    def __str__(self):
        return f"Player {self.name} has {len(self.all_cards)} cards."

    def remove_one(self):
        return self.all_cards.pop(0)

    def add_cards(self,new_card):
        if type(new_card) == list:
            self.all_cards.extend(new_card)
        else:
            self.all_cards.append(new_card)
```

```python
# SETUP THE GAME

player1 = Player("One")
player2 = Player("Two")

newdeck = Deck()
newdeck.shuffle()

for i in range(26):
    player1.add_cards(newdeck.draw_one())
    player2.add_cards(newdeck.draw_one())

game_on = True
```

```python
round_no = 0
war_count = 0

while game_on:

    # Round no counter

    round_no += 1
    print(f"Round {round_no}")

    # Check if someone has lost.

    if len(player1.all_cards) == 0:
        print("Player 1 is out of cards.")
        print("Player 2 has won!")
        print(f"Total number of wars: {war_count}")
        game_on = False
        break

    if len(player2.all_cards) == 0:
        print("Player 2 is out of cards.")
        print("Player 1 has won!")
        print(f"Total number of wars: {war_count}")
        game_on = False
        break

    # Putting a card on the table.

    player1_cards = []
    player2_cards = []

    player1_cards.append(player1.remove_one())
    player2_cards.append(player2.remove_one())

    compare = True
    

    while compare:

        if player1_cards[-1].value > player2_cards[-1].value:

            player1.add_cards(player1_cards)
            player1.add_cards(player2_cards)
            compare = False
            break

        elif player1_cards[-1].value < player2_cards[-1].value:

            player2.add_cards(player2_cards)
            player2.add_cards(player1_cards)
            compare = False
            break

        # ELSE IS THE WAR SITUATION
        else:
            print("WAR at Round {}".format(round_no))
            war_count += 1

            if len(player1.all_cards) < 3:
                print("Player 1 cannot declare WAR because there is not enough card!")
                print("Player 2 has won!")
                print(f"Total number of wars: {war_count}")
                game_on = False
                break

            elif len(player2.all_cards) < 3:
                print("Player 2 cannot declare WAR because there is not enough card!")
                print("Player 1 has won!")
                print(f"Total number of wars: {war_count}")
                game_on = False
                break

            else:
                for war_cards in range(3):
                    player1_cards.append(player1.remove_one())
                    player2_cards.append(player2.remove_one())
```
